B1. Use the following binning techniques to smooth the values of the "Age" attribute:

A. equi-width binning (3 bins).

```python
bin_number=3
```

```python
min_age=scoring['Age'].min()
max_age=scoring['Age'].max()
bin_width=(max_age-min_age)/ bin_number

bins = [min_age + i * bin_width for i in range(bin_number)]
bins.append(max_age) 

scoring['Age_Width_Bins'] = pd.cut(scoring['Age'], bins=bins, labels=['Bin 1', 'Bin 2', 'Bin 3'])

print(scoring[['Age', 'Age_Width_Bins']])
```

```python
scoring
```

B. equi-depth binning (3 bins).

```python
scoring_sorted = scoring.sort_values(by='Age')
```

```python
bin_edges = pd.qcut(scoring_sorted['Age'], q=bin_number, duplicates='drop', retbins=True)[1]

bin_labels = [f'Bin_{i+1}' for i in range(len(bin_edges) - 1)]

scoring_sorted['Age_Depth_Bins'] = pd.cut(scoring_sorted['Age'], bins=bin_edges, labels=bin_labels, include_lowest=True)

print(scoring_sorted[['Age', 'Age_Depth_Bins']])
```

B2. Use the following techniques to normalise the "Age" attribute:
    
 min-max normalization to transform the values onto the range [0.0-1.0].

 z-score normalization to transform the values

```python
scoring['Age_MinMax_Normalized'] = (scoring['Age'] - scoring['Age'].min()) / (scoring['Age'].max() - scoring['Age'].min())
```

```python
scoring['Age_ZScore_Normalized'] = (scoring['Age'] - scoring['Age'].mean()) / scoring['Age'].std()
```

```python
print(scoring[['Age', 'Age_MinMax_Normalized', 'Age_ZScore_Normalized']])
```

B3. Discretise the Age attribute into the following categories:

 Teenager = 1-16
 
 Young = 17-35
 
 Mid_Age = 36-55
 
 Mature = 56-70
 
 Old = 71+

```python
bins = [0, 16, 35, 55, 70, float('inf')]
labels = ['Teenager', 'Young', 'Mid_Age', 'Mature', 'Old']
```

```python
scoring['Age_Category'] = pd.cut(scoring['Age'], bins=bins, labels=labels, right=False)
category_counts = scoring['Age_Category'].value_counts()
```

```python
#Providing the frequency of each category in your data set.
print("Frequency of Each Category:")
print(category_counts)
```
